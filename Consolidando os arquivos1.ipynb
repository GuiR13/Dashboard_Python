```python
# importar bibliotecas

import pandas as pd 
import numpy as np

# ignorar alguns avisos
import warnings
warnings.filterwarnings('ignore')

# acessar componentes do sistema operacional, no caso para acessar a pasta lá encima
import os 
from pathlib import Path
```

```python
# Local diretório
local_diretorio = 'C:/Users/gui_r/Data_viking/Projeto_Dashboard_Evento/ajustados_2022'
```

```python
# como faço para ver todos os aquivos que estão ali dentro?
diretorio_path = Path(local_diretorio)

# é um tipo de uma pasta do Windows
type(diretorio_path)
```

```python
# listar arquivos
arquivos = diretorio_path.iterdir()

for loop in arquivos:
    print(loop)
```

```python
# sempre bom calcular o tamanho dos arquivos
# pegar o diretório do arquivo, passar para a biblioteca os, acesse esse diretório e calcule o tamanho desse arquivo

arquivos = diretorio_path.iterdir()

# função para acalcular o tamanho do arquivo
def bytes_mega(bytes):
    megabytes = bytes / (1024 * 1024)
    return megabytes

# total Mbs
total_mb = 0

for loop in arquivos:
    
    # buscando tamanho do arquivo
    tamanho = os.path.getsize(loop)
    
    # convertendo em mega
    conversao = bytes_mega( tamanho )
    
    # minerar informação de texto em python
    nome_arquivo = str(loop).split('\\')[-1]
    
    print(f' {nome_arquivo} : {conversao:.2f} MB')
    print('-' * 30)
    
    total_mb = total_mb + tamanho

print(f'Temos ao todo {bytes_mega(total_mb):.2f} Mbs de dados')
```

```python
# amostra
# ignorar alguns erros com: error_bad_lins=False
pd.read_csv(loop, error_bad_lines=False, sep=';', encoding='latin-1').head()
```

```python
# diretorio
arquivos = diretorio_path.iterdir()

# loop para consolidar os arquivos em 1 só arquivo
for contador,arq in enumerate(arquivos):
    print(contador, arq)
    
    # se for a 1ª interação, cria a base concolidada
    if contador == 0:
        # ler o arquivo do loop
        bade_dados = pd.read_csv(arq, error_bad_lines=False, sep=';', encoding='latin-1')
        # salvar na variável
        base_consolidada = bade_dados
        
    else:
        # ler os arquivos do loop
        base_dados = pd.read_csv(arq, error_bad_lines=False, sep=';', encoding='latin-1')
        # empilhar os dados
        base_consolidada = pd.concat( [base_consolidada, base_dados] )
        
# dimensão
base_consolidada.shape
```

```python
base_consolidada.tail()
```

```python
# exportação
#CSV
base_consolidada.to_csv('Base_Consolidada1.csv', index=False)
```

```python
# SQL
import sqlite3
```

```python
# criar o banco de dados

conexao = sqlite3.connect('Banco_Dados1.db')

# nome da tabela
nome_tabela = 'Base_Consolidada1'
```

```python
# enviar meus dados para esse banco de dados
# caso essa tabela exista, módulo append, caso não exista ele cria uma tabela nova
base_consolidada.to_sql(nome_tabela, conexao, if_exists='append', index=False)
```
